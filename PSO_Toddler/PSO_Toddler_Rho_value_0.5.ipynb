```python
import pandas as pd
import numpy as np
from pyswarm import pso

# Load the dataset
df = pd.read_csv("Autism-toddler.csv")

# Define the selected features
selected_features = ['A9', 'A7', 'A6', 'A5', 'A2']

# Create feature set and target set using the selected features
df = df[selected_features + ['Class/ASD Traits ']]

# Convert 'Class/ASD Traits' column to numeric
df['Class/ASD Traits '] = pd.to_numeric(df['Class/ASD Traits '], errors='coerce')

# Define the CES production function as the objective function
def ces_production_function(parameters, features):
    # Parameters are the weights for the selected features
    weights = parameters
    # Calculate the CES production function output based on the CES formula
    rho = 0.5  # You can adjust the value of rho as needed
    output = np.power(np.sum(np.power(weights * features, rho)), 1 / rho)
    return output

# Define severity categories based on Autism score
def categorize_severity(autism_score):
    if autism_score < 0.1:
        return 'Low Severity'
    elif autism_score < 0.3:
        return 'Moderate Severity'
    else:
        return 'High Severity'

# Iterate through rows and optimize for each row
optimal_weights_list = []
autism_score_list = []
severity_list = []

for index, row in df.iterrows():
    # Extract selected features and 'Class/ASD Traits' for the current row
    features = row[selected_features].values
    asd_traits = row['Class/ASD Traits ']

    # Number of parameters (weights for the selected features)
    num_parameters = len(selected_features)

    # Define bounds for the weights
    lower_bounds = [0.01] * num_parameters  # Set lower bounds to zero for non-negativity
    upper_bounds = [0.5] * num_parameters  # Adjust upper bounds as needed

    # Perform PSO optimization for the current row
    best_parameters, _ = pso(ces_production_function, lower_bounds, upper_bounds,
                             args=(features,), swarmsize=20, maxiter=100)

    # Calculate the Autism score based on the CES production function
    autism_score = ces_production_function(best_parameters, features)

    # Categorize the severity based on the Autism score
    severity = categorize_severity(autism_score)

    # Append the optimal weights, autism score, and severity to lists
    optimal_weights_list.append(best_parameters)
    autism_score_list.append(autism_score)
    #severity_list.append(severity)

# Convert the lists to DataFrames
optimal_weights_df = pd.DataFrame(optimal_weights_list, columns=[f'Weight_{feature}' for feature in selected_features])
autism_score_df = pd.DataFrame(autism_score_list, columns=['Autism_Score'])
severity_df = pd.DataFrame(severity_list, columns=['Severity'])

# Concatenate the DataFrames with the original dataset
result_df = pd.concat([df, optimal_weights_df, autism_score_df], axis=1)

# Save the result to a CSV file
result_df.to_csv('pso_toddler_rho_0.5.csv', index=False)  # Set index to False to exclude the index column
```
